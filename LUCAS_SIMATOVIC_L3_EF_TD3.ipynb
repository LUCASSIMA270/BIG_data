```python
# On importe tous les import dont on aura besoin pour la suite du TD

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
```

```python
# 1) On charge la base de données wine

dataset = load_wine()
X = dataset.data
y = dataset.target

# 2) On apprend les paramètres de la ACP à partir de la matrice X avec K valant 2 comosantes

pca = PCA(n_components=2)
pca.fit(X)

# 3) On projette les données de la matrice X pour obtenir une autre matrice Z

Z = pca.transform(X)
```

```python
# 4) On représente les données Z

plt.figure(figsize=(5, 5))

plt.scatter(Z[:, 0], Z[:, 1], c=y, cmap='cividis')
plt.xlabel('Première composante principale')
plt.ylabel('Deuxième composante principale')
plt.title('Représentatin des données de la matrice Z')
plt.colorbar(label='Appartenance du point à la classe')

plt.show()

```

```python
# 5) Calcul de la matrice de corrélation de X et Z

correlation_XZ = np.corrcoef(X.T, Z.T)

plt.figure(figsize=(5, 5))
plt.imshow(correlation_XZ, cmap='cividis')
plt.title('Matrice de correlation entre X et Z')
plt.colorbar()
plt.show()
```

```python
# Enfin nous pouvons calculer la matrice de correlation de X

Z = pca.transform(X)

# Nous transposons les matrices X et Z

X_transpose = X.T
Z_transpose = Z.T

# On calcule la matrice de correlation de X transposée

correlation_X = np.corrcoef(X_transpose)

# On calcule la matrice de correlation entre X et Z transposée

correlation_XZ = np.corrcoef(X_transpose, Z_transpose)

plt.figure(figsize=(5, 5))

plt.imshow(correlation_X, cmap='cividis')
plt.title('Transposée de la matrice de correlation de X')
plt.colorbar(label='Correlation')

plt.show()
```
